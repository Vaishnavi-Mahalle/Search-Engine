```python
from tkinter import *
from serpapi import GoogleSearch
import requests
```

```python
def getResults(search_query):
    params = {
  "q": search_query,
  "hl": "en",
  "gl": "in",
  "google_domain": "google.com",
  "api_key": "504075c023b0373199a46e996f510ab72e2f0f17bdbd4051b81ea51f8800ef3d"
}

    search = GoogleSearch(params)
    results = search.get_dict()
    answer=results["knowledge_graph"]["description"]
    return answer
```

```python
def getInfo():
    text_typed=input.get()
    try:
        answer=getResults(text_typed)  #It can happen that the API was not able to get the answer for the query question
    except:
        answer="Oops! no results found"
    result.config(text=answer)
    result.grid(column=0,row=4)
```

```python
window = Tk() # Defined a window
#window.geometry("600x600")
window.title("My Search Engine")
window.config(padx=10,pady=10,bg="#dfffff")

canvas=Canvas(width=600,height=420,bg="#dfffff",highlightthickness=0) #Define a canvas
logo_img=PhotoImage(file="Logo.png")  # Loaded the image file
canvas.create_image(300,200,image=logo_img)
canvas.grid(column=0,row=0)

# Input label
input = Entry(width=50) #Define a search box
input.grid(column=0,row=1)

# Blank canvas made of same color as background to insert a space 
canvas=Canvas(width=600,height=20,bg="#dfffff",highlightthickness=0)
canvas.grid(column=0,row=2)

# Search button
search_button=Button(text="Search" ,highlightthickness=0,width=20,command=getInfo,pady=5)
search_button.grid(column=0,row=3)

# Results label
result = Message(text="Here is the result",font=("Roboto", 10),width=600,bg="#dfffff")
result.grid_forget()

window.mainloop()
```
